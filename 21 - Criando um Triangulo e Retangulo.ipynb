```python
largura = 5
altura = 3

for i in range(altura):

    for j in range(largura):

        print("*", end = " ")

    print()


```

```python
contador = 1

for i in range (5):

    for j in range (1,contador + 1):

        print ("*")
        contador += 2

        if contador == 9:
            break
        else:
            print()
            



```
